# **Librerias**

```python
import matplotlib.pyplot as plt
import numpy as np
import random as rnd
```

***Función a generar con el método composición***


***Función de distribución de probabilidad***
$f(x) = \begin{cases}
\frac{1}{4} & \text{si } 0 \leq x \leq 2 \\
\frac{1}{2} & \text{si } 2 < x \leq 3 \\
0 & \text{en otro caso}
\end{cases} $


```python
def f(x):
    if 0 <= x <= 2:
        return (1/4)
    elif 2 < x <= 3:
        return (1/2)
    else:
        return 0

# Generar valores x
x = np.linspace(-1, 3, 400)

# Aplicar la función a cada valor de x
y = np.array([f(xi) for xi in x])

# Graficar la función
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.title('Gráfica de función de distribución de probabilidad')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

***Función de distribución acumulativa***
$F(x) = \begin{cases}
0 & \text{si } x < 0 \\
\frac{1}{4}x & \text{si } 0 \leq x \leq 2 \\
\frac{1}{2}x-\frac{1}{2} & \text{si } 2 < x \leq 3 \\
1 & \text{si } x > 3
\end{cases} $

```python
def f(x):
    if  x <= 0:
        return 0
    elif 0 <= x <= 2:
        return (0.25*x)
    elif 0 < x <= 3:
        return (0.5*x) -0.5
    else:
      return 1
# Generar valores x
x = np.linspace(-1, 5, 400)

# Aplicar la función a cada valor de x
y = np.array([f(xi) for xi in x])

# Graficar la función
plt.figure(figsize=(8, 6))
plt.plot(x, y,color="green")
plt.title('Gráfica de función de distribución acumulativa')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

# ***Aplicación del método de composición***

**Dividiendo la funcion en dos areas $A_{1}$ y $A_{2}$ y determinando la funcíon para cada una como $f_{1}(x)$ y $f_{2}(x)$ **

```python
#Función para área 1:
fig, a1x = plt.subplots()
a1x.fill_between([0, 2], [0.25,0.25],color="red")
plt.show()
```

Calculando $$A1 = b*h = 2*0.25 = 0.5$$ $$ f_{1}(x) = \frac{1}{4}$$

```python
#Función para área 2
fig, a2x = plt.subplots()
a2x.fill_between([2, 3], [0.5,0.5],color="green")
plt.show()
```

Calculando $$A2 = b*h = 1*0.5 = 0.5$$ $$ f_{2}(x) = \frac{1}{2}$$

La función general es:
$f(x) = A_{1}f_{1}+ A_{2}f_{2} = 0.5*\frac{1}{4} + 0.5*\frac{1}{2}$

***Histograma de la muestra***

```python
#Función que determina con que función hacer la evaluación
def aprox(n):
  #Lista que guarda los valores generados
  lista=[]
  for i in range(n):
    #Genera u1 ~ U(0,1)
    u1=0 + 3*rnd.random()
    if u1<=2:
      #Calcula valor de  x con f1
      x=0.25
    else:
      #Calcula valor de  x con f2
      x=0.5
    #Añade a la lista el valor de x
    lista.append(x)
  return lista
G=aprox(100000)
media=np.mean(G)
varianza=np.var(G)
print("La media es:",media)
print("La varianza es:",varianza)
```

```python
plt.hist(G,bins=500)
```

***Función de distribución acumulativa de la muestra***

```python
def cdf(muestra):
  n=len(muestra)
  x=np.sort(muestra)
  y=np.arange(0,n)/n
  return x,y
X,Y=cdf(G)
plt.step(X,Y,color="green")
plt.title('Gráfica de función de distribución acumulativa con la muestra')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
```
