```python
print("Ingresar primer numero de 3 digitos (semilla 1)")
x0=input()
x0=int(x0)
print("Ingresar segundo numero de 3 digitos (semilla 2)")
x1=input()
x1=int(x1)
print("Ingrese la cantidad de numeros aleatorios que desea")
n=input()
n=int(n)
Generados=[x0,x1]
for i in range(2,n):
  xn=int(x0*x1)
  xn=str(xn)
  xlis=list(xn)
  length=len(xlis)
  a=length%2
  # CICLO PARA AGREGAR UN 0 AL INICIO DE CADA NUMERO EN CASO DE QUE SEA IMPAR
  if a==0:
    xlis.insert(0,'0')
  else:
    xlis=xlis
  print(xlis)
  b=len(xlis)
  md=((b-1)/2)
  inte=md-1
  ext=int(md+2)
  x=xlis[inte:ext]
  x0=x1
  x1=''.join(x)
  Generados.append(x1)
  i=i+1
print("Los numeros generados por el metodo medio son:")
print(Generados)
```
