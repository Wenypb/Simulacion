```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
def aprox(n):
  lista=[]
  for i in range(n):
    u=rnd.random()
    x=((u**0.135)-(1-u)**0.135)/0.1975
    lista.append(x)
  return lista
G=aprox(10000000)
media=np.mean(G)
varianza=np.var(G)
print("La media es:",media)
print("La varianza es:",varianza)
```

```python
plt.hist(G)
```

```python
plt.plot(G)
```

```python
def cdf(muestra):
  n=len(muestra)
  x=np.sort(muestra)
  y=np.arange(0,n)/n
  return x,y
X,Y=cdf(G)
plt.step(X,Y)
```
