```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#MUESTRA DEL EXPERIMENTO
x=np.linspace(0,1,100)
u1=rnd.random()
u2=rnd.random()
plt.plot(x,np.sqrt(1-x**2))
plt.scatter(u1,u2,color="red")
plt.show()
```

```python
def funcion_pi_grafica(N):
  #LISTAS DE LOS VALORES DE U1 Y U2
  listau1=[]
  listau2=[]
  #CICLO PARA REPETIR EL EXPERIMENTO N VECES
    # n -> NUMERO DE EXITOS
  n=0
  for i in range(N):
    u1=rnd.random()
    u2=rnd.random()
    listau1.append(u1)
    listau2.append(u2)
    #CONDICION PARA EL EXITO
    x=np.linspace(0,1,100)
    plt.plot(x,np.sqrt(1-x**2))
    if u2<=np.sqrt(1-u1**2):
      plt.scatter(u1,u2,color="green")
      n=n+1
    else:
      plt.scatter(u1,u2,color="red")
  return n
```

```python
intentos=20
exitos=funcion_pi_grafica(intentos)
Estimador=(4*exitos)/intentos
print(Estimador)
```

```python
def funcion_pi(N):
  #LISTAS DE LOS VALORES DE U1 Y U2
  listau1=[]
  listau2=[]
  #CICLO PARA REPETIR EL EXPERIMENTO N VECES
    # n -> NUMERO DE EXITOS
  n=0
  for i in range(N):
    u1=rnd.random()
    u2=rnd.random()
    listau1.append(u1)
    listau2.append(u2)
    #CONDICION PARA EL EXITO
    if u2<=np.sqrt(1-u1**2):
      n=n+1
  return n
```
