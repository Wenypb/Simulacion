```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

# LECTURA

```python
# 2°
# Parámetros dados
media = 61.9
desviacion_estandar = 19 # Estimación basada en la distribución esperada
rango_1 = (0, 40)
rango_2 = (40, 60)
rango_3 = (60, 100)

# Número de muestras
n = 10000

# Crear una distribución normal con la media y desviación estándar
data = np.random.normal(media, desviacion_estandar, n)

# Calcular el porcentaje de datos en cada rango
percent_1 = np.sum((data >= rango_1[0]) & (data <= rango_1[1])) / n * 100
percent_2 = np.sum((data >= rango_2[0]) & (data <= rango_2[1])) / n * 100
percent_3 = np.sum((data >= rango_3[0]) & (data <= rango_3[1])) / n * 100

# Mostrar los porcentajes calculados
print(f"Porcentaje de Estudiantes con Prioridad:")
print(f"Alta: {percent_1:.2f}%")
print(f"Media: {percent_2:.2f}%")
print(f"Baja: {percent_3:.2f}%")

# Graficar la distribución
plt.figure(figsize=(8, 6))


# Agregar la distribución normal teórica para la media y desviación estándar
xmin, xmax = plt.xlim()
x = np.linspace(0, 100, 100)
p = norm.pdf(x, media, desviacion_estandar)
plt.plot(x, p, 'k', linewidth=2)

# Pintar los intervalos con diferentes colores
# Rango 1 (0 a 40)
plt.fill_between(x, 0, p, where=(x >= rango_1[0]) & (x <= rango_1[1]), color='red', alpha=0.3, label="Prioridad Alta")

# Rango 2 (40 a 60)
plt.fill_between(x, 0, p, where=(x > rango_1[1]) & (x <= rango_2[1]), color='blue', alpha=0.3, label="Prioridad Media")

# Rango 3 (60 a 100)
plt.fill_between(x, 0, p, where=(x > rango_2[1]) & (x <= rango_3[1]), color='purple', alpha=0.3, label="Prioridad Baja")

# Establecer los límites del eje X para que vaya solo de 0 a 100
plt.xlim(-5, 105)

# Etiquetas
plt.title("Distribución de Porcentaje de Aciertos 2° de Primaria")
plt.xlabel("Porcentaje")
plt.ylabel("Densidad de probabilidad")
plt.legend()

# Mostrar la gráfica
plt.show()
```
