# **Librerias**

```python
import matplotlib.pyplot as plt
import numpy as np
import random as rnd
```

***Función a generar con el método de la transformada inversa***
$$f(x)= -\frac{x}{2}+1$$ ***si*** $$0 \leq x \leq 2$$

***Función de distribución de probabilidad***
$f(x) = \begin{cases}
-\frac{x}{2}+1 & \text{si } 0 \leq x \leq 2 \\
0 & \text{en otro caso}
\end{cases} $


```python
def f(x):
    if 0 <= x <= 2:
        return (-x/2) + 1
    else:
        return 0

# Generar valores x
x = np.linspace(-1, 3, 400)

# Aplicar la función a cada valor de x
y = np.array([f(xi) for xi in x])

# Graficar la función
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.title('Gráfica de función de distribución de probabilidad')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

***Función de distribución acumulativa***
$F(x) = \begin{cases}
0 & \text{si } x < 0 \\
-\frac{x^2}{4}+x & \text{si } 0 \leq x \leq 2 \\
0 & \text{si } x > 2
\end{cases} $

```python
def f(x):
    if  x <= 0:
        return 0
    elif 0 <= x <= 2:
        return ((-x**2)/4) + x
    else:
      return 1
# Generar valores x
x = np.linspace(-1, 3, 400)

# Aplicar la función a cada valor de x
y = np.array([f(xi) for xi in x])

# Graficar la función
plt.figure(figsize=(8, 6))
plt.plot(x, y,color="green")
plt.title('Gráfica de función de distribución acumulativa')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

***Histograma de la muestra***

```python
def aprox(n):
  #Lista que guarda los valores generados
  lista=[]
  for i in range(n):
    #Genera u1 ~ U(0,1)
    u1=rnd.random()
    #Calcula valor de  x
    x=-2*((1-u1)**0.5)+2
    #Añade a la lista el valor de x
    lista.append(x)
  return lista
G=aprox(100000)
media=np.mean(G)
varianza=np.var(G)
print("La media es:",media)
print("La varianza es:",varianza)
```

```python
plt.hist(G,bins=500)
```

***Función de distribución acumulativa de la muestra***

```python
def cdf(muestra):
  n=len(muestra)
  x=np.sort(muestra)
  y=np.arange(0,n)/n
  return x,y
X,Y=cdf(G)
plt.step(X,Y,color="green")
plt.title('Gráfica de función de distribución acumulativa con la muestra')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
```
