```python
c=-0.3
S0=1000
sp.plotting.plot(S0*sp.exp(c*t), (t,0,10))
```

```python
S=sp.Function("S")
S(t)
ode=sp.Eq(sp.Derivative(S(t),t)-c*S(t),0)
ode
```

```python
sol=sp.dsolve(ode,S(t),ics={S(0):S0})
sol
sp.plotting.plot(sol.rhs)
```
