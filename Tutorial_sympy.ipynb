```python
import sympy as sp
sp.init_printing()
```

```python
S = sp.Symbol("S")
t = sp.Symbol("t")
c = sp.Symbol("c")
(t+4)**2
```

```python
((t+4)**2).expand()
```

```python
((t+4)**2).subs(t,5)
```

```python
sp.factor((t**2)+6*t+5)
```

```python
((t**2)+3*t+8*t+2).simplify()
```

```python
sp.plotting.plot(t*t)
```

```python
c=-0.3
S0=1000
sp.plotting.plot(S0*sp.exp(c*t), (t,0,10))
```

```python
S=sp.Function("S")
S(t)
ode=sp.Eq(sp.Derivative(S(t),t)-c*S(t),0)
ode
```

```python
sol=sp.dsolve(ode,S(t),ics={S(0):S0})
sol
sp.plotting.plot(sol.rhs)
```
