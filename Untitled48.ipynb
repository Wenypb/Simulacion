```python
import turtle

def dibujar_carro():
    # Configuración de la ventana
    turtle.bgcolor("skyblue")
    turtle.title("Carro de Hot Wheels")

    # Cuerpo del carro
    turtle.color("red")
    turtle.begin_fill()
    turtle.forward(200)  # Parte frontal
    turtle.left(90)
    turtle.forward(50)   # Altura del carro
    turtle.left(90)
    turtle.forward(200)  # Parte trasera
    turtle.left(90)
    turtle.forward(50)   # Altura del carro
    turtle.left(90)
    turtle.end_fill()

    # Ventana del carro
    turtle.color("black")
    turtle.begin_fill()
    turtle.left(90)
    turtle.forward(200)
    turtle.left(90)
    turtle.forward(30)
    turtle.left(90)
    turtle.forward(200)
    turtle.left(90)
    turtle.forward(30)
    turtle.end_fill()

    # Ruedas
    turtle.penup()
    turtle.goto(30, -20)
    turtle.pendown()
    turtle.color("black")
    turtle.begin_fill()
    turtle.circle(20)  # Rueda delantera
    turtle.end_fill()

    turtle.penup()
    turtle.goto(170, -20)
    turtle.pendown()
    turtle.begin_fill()
    turtle.circle(20)  # Rueda trasera
    turtle.end_fill()

    # Finalizar
    turtle.hideturtle()
    turtle.done()

# Llamar a la función para dibujar el carro
dibujar_carro()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Función para generar puntos en un cuadrado
def puntos_cuadrado(lado, num_puntos):
    x = np.linspace(0, lado, num_puntos)
    y = np.zeros(num_puntos)
    return np.concatenate((x, lado * np.ones(num_puntos), np.flip(x), np.zeros(num_puntos))), \
           np.concatenate((y, x, lado * np.ones(num_puntos), np.flip(y)))

# Función para generar puntos en un círculo
def puntos_circulo(radio, num_puntos):
    theta = np.linspace(0, 2 * np.pi, num_puntos)
    x = radio * np.cos(theta)
    y = radio * np.sin(theta)
    return x, y

# Configuración del gráfico
plt.figure(figsize=(12, 6))

# Pintar un cuadrado
x_cuadro, y_cuadro = puntos_cuadrado(5, 100)
plt.subplot(1, 2, 1)
plt.scatter(x_cuadro, y_cuadro, color='blue', s=10)  # Pintar el cuadrado
plt.xlim(-1, 6)
plt.ylim(-1, 6)
plt.title('Cuadrado con Puntos')
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(False)

# Pintar un círculo
x_circulo, y_circulo = puntos_circulo(3, 100)
plt.subplot(1, 2, 2)
plt.scatter(x_circulo, y_circulo, color='orange', s=10)  # Pintar el círculo
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.title('Círculo con Puntos')
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(False)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import tkinter as tk
from tkinter import simpledialog

class CarroDibujador:
    def __init__(self, master):
        self.master = master
        self.master.title("Dibujador de Carros")

        self.fig, self.ax = plt.subplots(figsize=(10, 5))
        self.ax.set_xlim(0, 6)
        self.ax.set_ylim(0, 4)
        self.ax.axis('off')

        self.boton_dibujar = tk.Button(master, text="Dibujar Carro", command=self.dibujar_carro)
        self.boton_dibujar.pack()

        self.boton_mostrar = tk.Button(master, text="Mostrar Gráfico", command=self.mostrar_grafico)
        self.boton_mostrar.pack()

        plt.ion()  # Activar modo interactivo

    def dibujar_carro(self):
        color = simpledialog.askstring("Color", "Introduce el color del carro (ej. red, blue):")
        if color:
            self.ax.clear()  # Limpiar la figura anterior

            # Cuerpo del carro
            carro_body = patches.PathPatch(
                patches.Path([(1, 1), (1, 2), (2, 3), (4, 3), (5, 2), (5, 1)],
                             [1, 1, 1, 1, 1, 79]), linewidth=1, edgecolor='black', facecolor=color)
            self.ax.add_patch(carro_body)

            # Techo del carro
            carro_roof = patches.Rectangle((2.5, 2), 1.5, 0.8, linewidth=1, edgecolor='black', facecolor='darkred')
            self.ax.add_patch(carro_roof)

            # Ventanas
            ventana1 = patches.Rectangle((2.6, 2.2), 0.5, 0.4, linewidth=1, edgecolor='black', facecolor='lightblue')
            ventana2 = patches.Rectangle((3.2, 2.2), 0.5, 0.4, linewidth=1, edgecolor='black', facecolor='lightblue')
            self.ax.add_patch(ventana1)
            self.ax.add_patch(ventana2)

            # Ruedas
            rueda1 = plt.Circle((1.5, 0.8), 0.25, color='black')
            rueda2 = plt.Circle((4.5, 0.8), 0.25, color='black')
            self.ax.add_artist(rueda1)
            self.ax.add_artist(rueda2)

            # Actualizar el gráfico
            self.ax.set_title('Carro Dibujado')
            self.ax.set_aspect('equal')
            self.ax.axis('off')

            plt.draw()

    def mostrar_grafico(self):
        plt.show()

if __name__ == "__main__":
    root = tk.Tk()
    app = CarroDibujador(root)
    root.mainloop()
```

```python
import turtle

def dibujar_llama(x, y):
    turtle.penup()
    turtle.goto(x, y)
    turtle.pendown()
    turtle.color("orange")
    turtle.begin_fill()

    # Dibuja la llama
    for _ in range(5):
        turtle.circle(15, 180)
        turtle.right(180)
    turtle.end_fill()

def dibujar_texto():
    turtle.penup()
    turtle.goto(-50, -20)
    turtle.pendown()
    turtle.color("white")
    turtle.write("HOT WHEELS", font=("Arial", 24, "bold"))

def dibujar_logo():
    # Configuración inicial
    turtle.bgcolor("blue")
    turtle.title("Logo de Hot Wheels")

    # Dibuja llamas
    dibujar_llama(-60, 0)
    dibujar_llama(-30, 10)
    dibujar_llama(0, 0)
    dibujar_llama(30, 10)
    dibujar_llama(60, 0)

    # Dibuja el texto
    dibujar_texto()

    # Finalizar
    turtle.hideturtle()
    turtle.done()

# Llamar a la función para dibujar el logo
dibujar_logo()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def llamachica(x_offset, y_offset, color):
    # Parámetros de la llama
    t = np.linspace(0, 2 * np.pi, 100)
    x = np.sin(t) ** 3
    y = (np.cos(t) - 0.5) ** 2

    # Ajustar la posición y el tamaño de la llama
    x = x * 0.2 + x_offset
    y = y * 0.2 + y_offset

    # Graficar la llama
    plt.fill(x, y, color=color)

def llamagrande(x_offset, y_offset, color):
    # Parámetros de la llama
    t = np.linspace(0, 2 * np.pi, 100)
    x = np.sin(t) ** 3
    y = (np.cos(t) - 0.5) ** 2

    # Ajustar la posición y el tamaño de la llama
    x = x * 0.5 + x_offset
    y = y * 0.5 + y_offset

    # Graficar la llama
    plt.fill(x, y, color=color)

# Crear el gráfico
plt.figure(figsize=(6, 8))
plt.xlim(-2, 2)
plt.ylim(-1, 2)

# Dibujar varias llamas
llamagrande(0, 0, 'red')
llamachica(0, 0, 'orange')
#llama(-0.2, 0.1, 'yellow')
#llama(0.1, 0.4, 'orange')

# Configuraciones finales
plt.title('Llamas Estilizadas')
plt.axis('off')  # Oculta los ejes

# Mostrar el gráfico
plt.show()
```
